Ejercicio 2: Estás trabajando en una empresa de comercio electrónico y tienes datos de ventas mensuales en relación con los gastos en publicidad en diferentes canales (por ejemplo, redes sociales, televisión, radio, etc.). Utiliza la regresión lineal para construir un modelo que relacione los gastos en publicidad con las ventas mensuales. Después, utiliza el modelo para predecir las ventas del próximo mes en función de los gastos en publicidad que se planean realizar.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
pd.pandas.set_option('display.max_columns', None)
pd.pandas.set_option('display.max_rows', None)
```

```python
# Generar datos sintéticos de horas de estudio
np.random.seed(0)

# Generar datos sintéticos de gastos en publicidad
gastos_publicidad = np.random.uniform(1000, 50000, size=1000).round(2)

# Generar datos sintéticos de ventas mensuales relacionados con los gastos en publicidad
ventas_mensuales = 10000 + 100 * gastos_publicidad + np.random.normal(0, 5000, size=1000).round(2)
```

```python
gastos_publicidad
```

```python
ventas_mensuales
```

Lo convertimos a un dataframe de pandas para poder manipularlo más fácilmente.

```python
# Crear un diccionario con los arrays
data = {'gastos_publicidad': gastos_publicidad, 'ventas_mensuales': ventas_mensuales}

# Crear el DataFrame a partir del diccionario
df = pd.DataFrame(data)

# Imprimir el DataFrame
df.head()
```

```python
# Crear el modelo de regresión lineal y entrenarlo
modelo = LinearRegression()
X = df['gastos_publicidad'].values.reshape(-1, 1)
Y = df['ventas_mensuales'].values
modelo.fit(X, Y)
```

```python
# Datos de alumno que estudio 5 horas

nueva_publi = 1000

# Predecir la calificación del alumno
nuevos_datos = np.array([[nueva_publi]])
precio_predicho = modelo.predict(nuevos_datos)

print('Venta predicha: ', precio_predicho)
```
